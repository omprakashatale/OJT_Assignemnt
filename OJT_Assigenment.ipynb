```python
import numpy as np
A = np.array([[5, 8, 9, 5, 0],
              [0, 1, 7, 6, 9],
              [2, 4, 5, 2, 4],
              [2, 4, 7, 7, 9],
              [1, 7, 0, 6, 9]], dtype=float)
print("Original Matrix A:")
print(A)
rows, cols = A.shape
for i in range(min(rows, cols)):
    pivot_row = i
    while pivot_row < rows and A[pivot_row, i] == 0:
        pivot_row += 1

    if pivot_row == rows:
        continue
    A[[i, pivot_row]] = A[[pivot_row, i]]

    A[i] /= A[i, i]

    for j in range(i + 1, rows):
        A[j] -= A[j, i] * A[i]

print("\nRow Echelon Form of Matrix A:")
print(A)
```

```python
#Q4
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema

np.random.seed(1)
coefficients = np.random.randint(-10, 10, size=6)
def polynomial(x):
    return np.polyval(coefficients, x)

# Generate x values
x_values = np.linspace(-5, 5, 400)

# Calculate y values using the polynomial function
y_values = polynomial(x_values)

# Plot the polynomial
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Polynomial', color='blue')
plt.title('Plot of a 5th Degree Polynomial')
plt.xlabel('x')
plt.ylabel('y')

# Find local maxima and minima
maxima_indices = argrelextrema(y_values, np.greater)[0]
minima_indices = argrelextrema(y_values, np.less)[0]

# Annotate maxima and minima
for index in maxima_indices:
    plt.annotate('Maxima', xy=(x_values[index], y_values[index]), xytext=(-20, 20), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->', color='red', connectionstyle='arc3,rad=.5'))
for index in minima_indices:
    plt.annotate('Minima', xy=(x_values[index], y_values[index]), xytext=(-20, -20), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->', color='green', connectionstyle='arc3,rad=.5'))

plt.grid(True)
plt.legend()
plt.show()
```
