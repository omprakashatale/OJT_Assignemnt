```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
iris_df = pd.read_csv("iris.csv")

print(iris_df.head())

summary_stats = iris_df.describe()
print("\nSummary Statistics:")
print(summary_stats)

iris_df.hist(figsize=(10, 8))
plt.suptitle("Histograms of Numerical Columns")
plt.tight_layout()
plt.show()

iris_df.plot(kind='box', figsize=(10, 6), subplots=True, layout=(2, 2), sharex=False, sharey=False)
plt.suptitle("Box plots of Numerical Columns")
plt.tight_layout()
plt.show()

sns.pairplot(iris_df, diag_kind='kde')
plt.suptitle("Pairwise Scatter plots with KDE Diagonal")
plt.show()
```
